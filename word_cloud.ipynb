```python
# pip install wordcloud
# Type the above command in Terminal to
# Installing the packages.

# importing the packages
from wordcloud import WordCloud, ImageColorGenerator
import numpy as np
import matplotlib.pyplot as plt
```

```python
text = ''

# Opening the file. and encoded with the UTF-8.
with open('F:/basic_python/jupyter content/practice hub/summary_of_chapter1.txt', encoding = 'utf-8') as f:
    text = ''.join(f.readlines()) # read the lines.
```

```python
# creating the object for wordcloud and keep background as white.
wc = WordCloud(background_color='white') # keeping background color as white for the image.
# NOTE: for the parameters, review the Doc.

wc.generate(text) # It collects the unique words from the text file by neglating the Conjuntions, preposition and somemore
# grammatical notion.

wc.to_file('new_image.jpg') # saves the file into JPG format.

plt.figure(figsize=(12,9)) # Fixing the Image size. 

plt.imshow(wc) # showing the Image.

plt.axis('off') # disabling the axis.
print('End result of the Program.')
```
